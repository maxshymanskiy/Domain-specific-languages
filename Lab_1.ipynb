**Завдання 1**: Напишіть програму, яка обчислює суму всіх парних чисел у
заданому діапазоні.
1. Створіть дві цілочесельні змінні:
***var_start*** та ***var_end***. Надайте їм значення, що визначають початок та кінець діапазону
відповідно.
2. Використайте цикл ***for*** для ітерації по числах у цьому діапазоні.
3. У змінну ***total_sum*** зберегти суму парних чисел із вказаного діапазону, а
також вивести резутат на екран.

```python
var_start = 1
var_end = 8
total_sum = 0

for i in range(var_start, var_end):
    if i % 2 == 0:
        total_sum += i
        
print(f"Total sum of even numbers: {total_sum}")
```

**Завдання 4**: Напишіть програму, яка визначає придатність студента до
стипендії на основі його академічних показників та активності.
1. Створіть три змінні:
- ***gpa*** типу *float* (від 2 до 5)
- ***is_active_in_student_council*** типу *boolean* та
- ***has_volunteer_experience*** типу *boolean*.
2. Потрібно перевірити наступні умови, а результат зберегти у змінну
***scholarship_type***, яка може набувати значень ***'full'***, ***'partial'*** або ***'no'***:
- Студент отримує повну стипендію, якщо його ***gpa*** більше або
дорівнює 4.5 і він є активним у студентській раді.
- Студент отримує часткову стипендію, якщо його ***gpa*** більше
або дорівнює 4.0 або він має досвід волонтерства.
- Студент не отримує стипендію, якщо жодна з вищезазначених
умов не виконується. Використайте оператор ***not*** для перевірки
цієї умови.
3. Запусть даний код декілька разів із різними вхідними даними, що
продемонструвати виконання кожної з умов.

```python
gpa = 4.6
is_active_in_student_council = False
has_volunteer_experience = False

def determine_scholarship_type(gpa, is_active, has_exp):
    if not (2.0 <= gpa <= 5.0):
        return "invalid gpa"
    
    if gpa >= 4.5 and is_active:
        return "full"
    elif gpa >= 4.0 or has_exp:
        return "partial"
    elif not (gpa >= 4.5 and is_active) and not (gpa >= 4.0 or has_exp):
        return "no"

scholarship_type = determine_scholarship_type(
    gpa, is_active_in_student_council, 
    has_volunteer_experience
)

print(f"GPA: {gpa}")
print(f"Active in student council: {is_active_in_student_council}")
print(f"Expirience in volunteering: {has_volunteer_experience}")
print(f"Result: {scholarship_type}")
```

**Завдання 5**: Напишіть програму, яка здійснює певні арифметичні операції з
числами у заданому діапазоні, пропускаючи числа, які діляться на 3 або 5, за
наступними правилами:
- Якщо число парне, то до загальної суми потрібно додати квадрат
цього числа.
- Якщо число не парне, то ми віднімаємо його від загальної суми.
1. Створіть дві змінні: ***var_start*** та ***var_end***. Присвойте їм цілі числові
значення, що визначають початок та кінець діапазону.
2. Використайте цикл ***for*** для ітерації по числах у цьому діапазоні.
3. Всередині циклу реалізуйте вищеперераховані правила та збережіть
результат у змінну ***total_result***.
4. Після завершення циклу виведіть на екран кінцеве значення ***total_result***.

```python
var_start = 1
var_end = 22
total_result = 0

for i in range(var_start, var_end):
    if (i % 3 == 0) or (i % 5 == 0):
        continue
        
    if i % 2 == 0:
        total_result += i**2
    else:
        total_result -= i

print(f"Result: {total_result}")
```

**Завдання 7**: Напишіть програму, яка шукає "щасливе" число у заданому
діапазоні.
"Щасливе" число визначається як таке, що ділиться на 7, не ділиться на 3
і є парним.
Додаткові умови:
- Не парні числа не повинні опрацьовуватись.
- Цикл має зупинитися, як тільки буде знайдено перше таке число,
яке відповідатиме статусу "Щасливе" та виведене на екран з
відповідним повідомленням.
- Цикл ***for*** заборонено використовувати.
- Створіть змінні ***start_number*** та ***end_number*** для визначення
діапазону.
- Якщо "Щасливе" число у вибраному діапазоні не буде знайдено,
то вивести відповідне повідомлення.

```python
start_number = 5
end_number = 100
is_found = False
i = start_number

while i < end_number:
    if i % 2 == 0:
        if (i % 7 == 0) and not (i % 3 == 0):
            print(f"Your lucky number is {i}")
            is_found = True
            break
            
    i += 1

if not is_found:
    print("Not found")
```
