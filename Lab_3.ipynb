### 3\. –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –¥—É–±–ª—ñ–∫–∞—Ç—ñ–≤ —Ç–∞ —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è

3.1. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥, –≤–∏–∑–Ω–∞—á—Ç–µ, —Å–∫—ñ–ª—å–∫–∏ —Ä–∞–∑—ñ–≤ —É —Å–ø–∏—Å–∫—É **`processed_inventory`** –∑—É—Å—Ç—Ä—ñ—á–∞—î—Ç—å—Å—è —Ç–æ–≤–∞—Ä "Laptop". –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

3.2. –í–∏–∫–æ–Ω–∞–π—Ç–µ —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Å–ø–∏—Å–∫—É **`processed_inventory`** –∑–∞ –∞–ª—Ñ–∞–≤—ñ—Ç–æ–º, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥. –í–∏–≤–µ–¥—ñ—Ç—å –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π —Å–ø–∏—Å–æ–∫.

### 4\. –í–∏–ª—É—á–µ–Ω–Ω—è —Ç–æ–≤–∞—Ä—ñ–≤

4.1. –í–∏–¥–∞–ª—ñ—Ç—å –∑ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–æ–≥–æ —Å–ø–∏—Å–∫—É –ø–µ—Ä—à–∏–π —Ç–æ–≤–∞—Ä, —è–∫–∏–π –º–∞—î –Ω–∞–∑–≤—É "Mouse", –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥.

4.2. –í–∏–¥–∞–ª—ñ—Ç—å —Ç–æ–≤–∞—Ä —ñ–∑ –∑–∞–¥–∞–Ω–∏–º —ñ–Ω–¥–µ–∫—Å–æ–º (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –µ–ª–µ–º–µ–Ω—Ç –Ω–∞ –ø–æ–∑–∏—Ü—ñ—ó 3) –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ **`del`**.

````python
inventory = [
"Laptop:15",
"Monitor:25",
"Keyboard:50",
"Mouse:45",
"Webcam:10",
"Headphones:30",
"Laptop:5"
]

##### 1.1
processesd_inventory = list()
for item in inventory:
    # 1.2
    item_name = item.split(":")[0]
    processesd_inventory.append(item_name)
# 1.3
print(f"–°–ø–∏—Å–æ–∫ –Ω–∞–∑–≤ —Ç–æ–≤–∞—Ä—ñ–≤ (–Ω–æ—Ä–º–∞–ª—ñ–∑–∞—Ü—ñ—è): {processesd_inventory}")
print("\n" + 55 * "=" + "\n")

##### 2.1
processesd_inventory.append("Speaker")

# 2.2
processesd_inventory.extend(["Printer", "Scanner"])
print(f"–ü—ñ—Å–ª—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è –Ω–æ–≤–∏—Ö —Ç–æ–≤–∞—Ä—ñ–≤: {processesd_inventory}")
print("\n" + 55 * "=" + "\n")

##### 3.1
product_name = "Laptop"
product_count = processesd_inventory.count(product_name)
print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤—Ö–æ–¥–∂–µ–Ω—å '{product_name}': {product_count}")

# 3.2
processesd_inventory.sort()
print(f"–í—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π —Å–ø–∏—Å–æ–∫: {processesd_inventory}")
print("\n" + 55 * "=" + "\n")

##### 4.1
processesd_inventory.remove("Mouse")
# 4.2
del processesd_inventory[3] 
print(processesd_inventory)
````

### –í—Ö—ñ–¥–Ω—ñ –¥–∞–Ω—ñ (–ü–æ—á–∞—Ç–∫–æ–≤–∏–π —Å–ø–∏—Å–æ–∫)

### 1\. –î–æ–¥–∞–≤–∞–Ω–Ω—è —Ç–∞ –≤—Å—Ç–∞–≤–∫–∞

### 3\. –£–º–æ–≤–Ω–∏–π –∞–Ω–∞–ª—ñ–∑ —Ç–∞ –≤–∏–¥–∞–ª–µ–Ω–Ω—è –∑–∞ —ñ–Ω–¥–µ–∫—Å–æ–º 

3.1. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ —Ü–∏–∫–ª —Ç–∞ —É–º–æ–≤–Ω–∏–π –æ–ø–µ—Ä–∞—Ç–æ—Ä, —Å—Ç–≤–æ—Ä—ñ—Ç—å –Ω–æ–≤–∏–π —Å–ø–∏—Å–æ–∫ **`high_grades`**, —è–∫–∏–π –º—ñ—Å—Ç–∏—Ç–∏–º–µ –ª–∏—à–µ –æ—Ü—ñ–Ω–∫–∏, –±—ñ–ª—å—à—ñ –∞–±–æ —Ä—ñ–≤–Ω—ñ **`90`**.

3.2. –ó—ñ —Å–ø–∏—Å–∫—É **`grades`** –≤–∏–¥–∞–ª—ñ—Ç—å –Ω–∞–π–Ω–∏–∂—á—É –æ—Ü—ñ–Ω–∫—É, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥ –≤–∏–¥–∞–ª–µ–Ω–Ω—è –∑–∞ —ñ–Ω–¥–µ–∫—Å–æ–º, —è–∫–∏–π –¥–æ–∑–≤–æ–ª—è—î –∑–±–µ—Ä–µ–≥—Ç–∏ –∑–Ω–∞—á–µ–Ω–Ω—è –≤–∏–¥–∞–ª–µ–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç—É —Ç–∞ –≤–∏–≤–µ—Å—Ç–∏ —Ü–µ –∑–Ω–∞—á–µ–Ω–Ω—è –Ω–∞ –¥—Ä—É–∫.

### 4\. –°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Ç–∞ –∑—Ä—ñ–∑–∏ 

4.1. –°—Ç–≤–æ—Ä—ñ—Ç—å –Ω–æ–≤–∏–π —Å–ø–∏—Å–æ–∫ **`top_three`**, —è–∫–∏–π –º—ñ—Å—Ç–∏—Ç–∏–º–µ —Ç—Ä–∏ –Ω–∞–π–≤–∏—â—ñ –æ—Ü—ñ–Ω–∫–∏, –≤—Ä–∞—Ö–æ–≤—É—é—á–∏ —Ç–µ, —â–æ –≤—Ö—ñ–¥–Ω–∏–π —Å–ø–∏—Å–æ–∫ –∑–∞–±–æ—Ä–æ–Ω–µ–Ω–æ –º–æ–¥–∏—Ñ—ñ–∫—É–≤–∞—Ç–∏. –í–∏–≤–µ–¥—ñ—Ç—å **`top_three`**.

````python
grades = [92, 78, 85, 92, 63, 100, 88, 78, 95, 81, 78]

##### 1.1
grades.append(75)

# 1.2
grades.insert(2, 100)
print(f"–°–ø–∏—Å–æ–∫ –ø—ñ—Å–ª—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è –æ—Ü—ñ–Ω–æ–∫: {grades}")
print("\n" + 55 * "=" + "\n")

##### 2.1 
grade = 78
grade_count = grades.count(grade)
print(f"–û—Ü—ñ–Ω–∫–∞ {grade} –∑—É—Å—Ç—Ä—ñ—á–∞—î—Ç—å—Å—è: {grade_count} —Ä–∞–∑(–∏)")

# 2.2
grades.remove(92)
print(grades)
print("\n" + 55 * "=" + "\n")

##### 3.1
high_grades = list()
for gr in grades:
    if gr >= 90:
        high_grades.append(gr)
print(f"–û—Ü—ñ–Ω–∫–∏ >= 90: {high_grades}")

# 3.2
lowest_grade_index = grades.index(min(grades))
print(f"–í–∏–¥–∞–ª–µ–Ω–∞ –Ω–∞–π–Ω–∏–∂—á–∞ –æ—Ü—ñ–Ω–∫–∞: {grades.pop(lowest_grade_index)}.")
print(grades)
print("\n" + 55 * "=" + "\n")

##### 4.1 
grades_copy = grades[:]
grades_copy.sort(reverse=True)
top_three = grades_copy[:3]
print(f"–¢—Ä–∏ –Ω–∞–π–≤–∏—â—ñ –æ—Ü—ñ–Ω–∫–∏: {top_three}")
````

1.3. –î–ª—è —Å–ø—ñ–≤—Ä–æ–±—ñ—Ç–Ω–∏–∫–∞ **`"E104"`** –∑–±—ñ–ª—å—à—Ç–µ –∑–∞—Ä–ø–ª–∞—Ç—É (`salary`) –Ω–∞ **`5000`**.

### 2\. –ê–Ω–∞–ª—ñ–∑ —É–Ω—ñ–∫–∞–ª—å–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å (–ú–Ω–æ–∂–∏–Ω–∏) 

2.1. –°—Ç–≤–æ—Ä—ñ—Ç—å –º–Ω–æ–∂–∏–Ω—É (**`set`**) **`all_departments`**, —è–∫–∞ –º—ñ—Å—Ç–∏—Ç–∏–º–µ —É–Ω—ñ–∫–∞–ª—å–Ω—ñ –Ω–∞–∑–≤–∏ –≤—ñ–¥–¥—ñ–ª—ñ–≤ —É—Å—ñ—Ö —Å–ø—ñ–≤—Ä–æ–±—ñ—Ç–Ω–∏–∫—ñ–≤.

### 3\. –í–∏–¥–∞–ª–µ–Ω–Ω—è —Ç–∞ —Ñ—ñ–ª—å—Ç—Ä–∞—Ü—ñ—è 

3.1. –í–∏–¥–∞–ª—ñ—Ç—å —Å–ø—ñ–≤—Ä–æ–±—ñ—Ç–Ω–∏–∫–∞ –∑ ID **`"E103"`** –∑—ñ —Å–ª–æ–≤–Ω–∏–∫–∞ **`employees`**, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥.
**
3.2. –°—Ç–≤–æ—Ä—ñ—Ç—å –Ω–æ–≤–∏–π —Å–ø–∏—Å–æ–∫ **`high_earners`**, —è–∫–∏–π –º—ñ—Å—Ç–∏—Ç–∏–º–µ –∫–æ—Ä—Ç–µ–∂—ñ (`tuple`) `(—ñ–º'—è, –∑–∞—Ä–ø–ª–∞—Ç–∞)` —Ç–∏—Ö —Å–ø—ñ–≤—Ä–æ–±—ñ—Ç–Ω–∏–∫—ñ–≤, –∑–∞—Ä–ø–ª–∞—Ç–∞ (`salary`) —è–∫–∏—Ö –ø–µ—Ä–µ–≤–∏—â—É—î **`80000`**.

### 4\. –°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è —Ç–∞ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è 

4.1. –°—Ç–≤–æ—Ä—ñ—Ç—å –Ω–æ–≤–∏–π —Å–ø–∏—Å–æ–∫ **`sorted_by_name`**, —â–æ –º—ñ—Å—Ç–∏—Ç–∏–º–µ ID —Å–ø—ñ–≤—Ä–æ–±—ñ—Ç–Ω–∏–∫—ñ–≤, –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏—Ö –∑–∞ —ó—Ö–Ω—ñ–º —ñ–º'—è–º (`name`) –≤ –∞–ª—Ñ–∞–≤—ñ—Ç–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é **`sorted()`** –∑—ñ —Å–ø–µ—Ü—ñ–∞–ª—å–Ω–æ—é **lambda-—Ñ—É–Ω–∫—Ü—ñ—î—é** —è–∫ –∫–ª—é—á–µ–º —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è. –í–∏–≤–µ–¥—ñ—Ç—å —Ü–µ–π —Å–ø–∏—Å–æ–∫.

4.2. –í–∏–≤–µ–¥—ñ—Ç—å –∑–∞–≥–∞–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Å–ø—ñ–≤—Ä–æ–±—ñ—Ç–Ω–∏–∫—ñ–≤ —É –∫–æ–∂–Ω–æ–º—É –≤—ñ–¥–¥—ñ–ª—ñ. (*–ü—ñ–¥–∫–∞–∑–∫–∞: –¥–ª—è —Ü—å–æ–≥–æ –º–æ–∂–Ω–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏ —Å–ª–æ–≤–Ω–∏–∫, —â–æ –ø—ñ–¥—Ä–∞—Ö–æ–≤—É—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤—Ö–æ–¥–∂–µ–Ω—å, –∞–±–æ –ø–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ –≤ —Ü–∏–∫–ª—ñ, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ —Ä–∞–Ω—ñ—à–µ —Å—Ç–≤–æ—Ä–µ–Ω—É –º–Ω–æ–∂–∏–Ω—É –≤—ñ–¥–¥—ñ–ª—ñ–≤*).

````python
employees = {
    "E101": {"name": "Alice Johnson", "department": "Marketing", "salary": 65000, "skills": ["SEO", "Content_Creation"]},
    "E102": {"name": "Bob Smith", "department": "IT", "salary": 85000, "skills": ["Python", "Databases", "Networking"]},
    "E103": {"name": "Charlie Brown", "department": "Marketing", "salary": 70000, "skills": ["Social_Media", "Analytics"]},
    "E104": {"name": "Diana Prince", "department": "IT", "salary": 92000, "skills": ["Cloud", "Python", "Security"]}
}

##### 1.1 
employees["E105"] = {
    "name": "Eve Adams",
    "department": "Finance", 
    "salary": 75000,
    "skills": ["Budgeting", "Reporting"]
}
print("–ü—ñ—Å–ª—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è E105:")
print(employees)

# 1.2
employees['E102']['department'] == "Development"
print(f"\nE102 –ø—ñ—Å–ª—è –∑–º—ñ–Ω–∏ –≤—ñ–¥–¥—ñ–ª—É: {employees['E102']}")

# 1.3
employees['E104']['salary'] += 500
print(f"E104 –ø—ñ—Å–ª—è –ø—ñ–¥–≤–∏—â–µ–Ω–Ω—è –∑–∞—Ä–ø–ª–∞—Ç–∏: {employees['E104']}")
print("\n" + 55 * "=" + "\n")

##### 2.1
all_departments = set()
for employee in employees.values():
    dept = employee["department"]
    all_departments.add(dept)
print(f"–£–Ω—ñ–∫–∞–ª—å–Ω—ñ –≤—ñ–¥–¥—ñ–ª–∏: {all_departments}")

# 2.2
core_skills = set()
for emp in employees.values():
    for skill in emp["skills"]:
        core_skills.add(skill)

core_skills = list(core_skills)
print(f"–£–Ω—ñ–∫–∞–ª—å–Ω—ñ –Ω–∞–≤–∏—á–∫–∏: {core_skills}")
print("\n" + 55 * "=" + "\n")

##### 3.1
print(f"–í–∏–¥–∞–ª–µ–Ω–∏–π —Å–ø—ñ–≤—Ä–æ–±—ñ—Ç–Ω–∏–∫: {employees.pop("E103")}")

# 3.2
high_earners = list()
for emp in employees.values():
    if emp['salary'] > 80000:
        high_earners.append((emp['name'], emp['salary']))
print(f"–°–ø—ñ–≤—Ä–æ–±—ñ—Ç–Ω–∏–∫–∏ –∑ –∑–∞—Ä–ø–ª–∞—Ç–æ—é > 80000: {high_earners}")
print("\n" + 55 * "=" + "\n")

##### 4.1
sorted_by_name = sorted(employees.keys(), key=lambda x: employees[x]['name'])
print(f"ID –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω—ñ –∑–∞ —ñ–º–µ–Ω–µ–º: {sorted_by_name}")

# 4.2
department_count = {}
for emp in employees.values():
    dept = emp['department']
    if dept in department_count:
        department_count[dept] += 1
    else:
        department_count[dept] = 1
print("–ö—ñ–ª—å–∫—ñ—Å—Ç—å —Å–ø—ñ–≤—Ä–æ–±—ñ—Ç–Ω–∏–∫—ñ–≤ –ø–æ –≤—ñ–¥–¥—ñ–ª–∞—Ö:")
for dept, count in department_count.items():
    print(f"{dept}: {count}")
````
