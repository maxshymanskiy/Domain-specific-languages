## Завдання 1. Робота з об'єктами DataFrame

**Вхідні дані:** Файл `nfl.csv` містить список гравців Національної ліги з наступними стовпцями: `Name`, `Team`, `Position`, `Birthday` та `Salary`.

**1) Імпортуйте файл `nfl.csv` в об'єкт `nfl` DataFrame**
- Забезпечте збереження інформації стовпця `Birthday` у типі дата/час.
- Реалізуйте створення користувацького індексу об'єкта DataFrame для збереження імен гравців.

```python
import pandas as pd

nfl = pd.read_csv("csv_files/nfl.csv", parse_dates=["Birthday"], index_col="Name")
print(nfl)
```

**2) Здійсніть наступні маніпуляції з імпортованими даними:**
- Підрахуйте кількість гравців по командам в цьому наборі даних,
збережіть результат в об'єкті `team_capacity`.
- Виведіть перших та останніх три елемента об'єкта `team_capacity`.
- Виведіть інформацію з об'єкта `nfl` про найбільшоплачуваного старшого
гравця команди, яка має найбільшу чисельністю гравців.
- Виведіть інформацію з об'єкта `nfl` про найбільшоплачуваного молодшого
гравця команди, яка має найменшу чисельністю гравців.

```python
team_capacity = nfl['Team'].value_counts()

print(team_capacity.head(3))
print(team_capacity.tail(3))

max_team = team_capacity.idxmax()
team_largest = nfl[nfl['Team'] == max_team]
highest_paid_in_largest = team_largest.nlargest(1, 'Salary')
print(highest_paid_in_largest[['Team', 'Position', 'Birthday', 'Salary']])

min_team = team_capacity.idxmin()
team_smallest = nfl[nfl['Team'] == min_team]
highest_paid_in_smallest = team_smallest.nlargest(1, 'Salary')
print(highest_paid_in_smallest[['Team', 'Position', 'Birthday', 'Salary']])
```

**3) Здійсніть наступні маніпуляції з імпортованими даними:**
- Зробити стовпець Team об'єкта nfl новим індексом об'єкта DataFrame.
- Переіменувати стовпець “Birthday” у “Date of Birth”.
- Виведіть інформацію для діапазону даних від 10 до 20 рядка за
виключенням стовпців типу дата/час.

```python
new_nfl = nfl.set_index('Team')
new_nfl = new_nfl.rename(columns={"Birthday": "Date of Birth"})
print(new_nfl.iloc[10:21].drop(columns=["Date of Birth"]))
```

## Завдання 2. Робота з об’єктами DataFrame

**Вхідні дані:** `Netflix.csv` містить колекцію приблизно із 6000 фільмів та серіалів. Він включає чотири стовпці: назву відео `(title)`, режисера `(director)`, дату появи фільму у списку `Netflix (Date_added)` та тип/категорію `(type)`.

**1) Добийтесь оптимального використання цим набором даних пам’яті та
зручності представлення даних:**

```python
netflix = pd.read_csv("csv_files/Netflix.csv", parse_dates=["date_added"])
netflix['type'] = netflix['type'].astype('category')
print(netflix.info())
```

**2) Знайдіть усі рядки з назвою Limitless:**

```python
limitless = netflix["title"] == "Limitless"
print(netflix[limitless])
```

**3) Знайдіть всі рядки з режисером Robert Rodriguez та типим Movie:**

```python
is_robert = netflix["director"] == "Robert Rodriguez"
is_movie = netflix["type"] == "Movie"

netflix[is_movie & is_robert]
```

**4) Знайти усі рядки з датою початку трансляції на Netflix `2019-07-31` або
режисером `Rober Altman`.**

```python
is_date = netflix["date_added"] == "2019-07-31"
is_altman = netflix["director"] == "Robert Altman"
netflix[is_date | is_altman]
```

**5) Знайдіть усі рядки з рижесером `Orson Welles`, `Aditya Kripalani` та `Sam Raimi` не використовуючи логічний оператор `OR ( | )`:**

```python
list_of_directors = ["Orson Welles", "Aditya Kripalani", "Sam Raimi"]
all_rows = netflix["director"].isin(list_of_directors)
netflix[all_rows]
```

**6) Знайдіть усі рядки зі значеннями стовпця `date_added` між 1 травня та 1 липня 2019 не використовуючи логічний оператор `AND ( & )`:**

```python
dates_range = netflix["date_added"].between("2019-05-01", "2019-07-01")
netflix[dates_range]
```

**7) Видаліть всі рядки `NaN` в стовпці `director`. Виведіть перші 5 рядків після операції видалення пустих значень.**

```python
cleared_netflix = netflix.dropna(subset=["director"])
cleared_netflix.head(5)
```

**8) Знайдіть всі дні, в які `Netflix` додав в каталог лише одну картину:**

```python
date_counts = cleared_netflix['date_added'].value_counts()
dates_with_one = date_counts[date_counts == 1].index
result = cleared_netflix[cleared_netflix['date_added'].isin(dates_with_one)]
result
```
